```python
# statistiche locali

x = np.float64(io.imread(path+'filamento.jpg'))

local_mean = ndi.generic_filter(x, np.mean, (3,3))
local_dev = ndi.generic_filter(x, np.std, (3,3))

mean = np.mean(x)
dev = np.std(x)

mask = (local_mean <= 0.4 * mean) & (local_dev >= 0.02*dev) & (local_dev < 0.4*dev)
y = 4*x*mask + (1-mask)*x

plt.close('all')
plt.figure(1)
plt.imshow(x, clim=[0,255], cmap='gray')

plt.figure(2)
plt.imshow(mask, clim=[0,1], cmap='gray')

plt.figure(3)
plt.imshow(y, clim=[0,255], cmap='gray')
```

```python
# operazioni geometriche 

path_im = path+'lena.jpg'
x = np.float64(io.imread(path_im))

#decimazione
y = x[::5,::5]
plt.figure(1)
plt.subplot(1,2,1)
plt.imshow(x, clim=[0,255], cmap='gray')
plt.subplot(1,2,2)
plt.imshow(y, clim=[0,255], cmap='gray')
plt.show()

#ridimensionamento 
y = rescale(x, 2, order=1)
plt.figure(2)
plt.subplot(1,2,1)
plt.imshow(x, clim=[0,255], cmap='gray')
plt.subplot(1,2,2)
plt.imshow(y, clim=[0,255], cmap='gray')
plt.show()
```

```python
# trasformazione geometrica

path_im = path+'lena.jpg'
x = np.float32(io.imread(path_im))
print(f"dimensioni prima della trasformazione:{x.shape}\tdopo la trasformazione {x[252:277, 240:290].shape}")
x = x[252:277, 240:290]
M,N = x.shape

plt.figure(1)
plt.imshow(x, clim=[0,255], cmap='gray')

# è possibile definire un'operazione geometrica a partire dalla matrice di trasformazione

A = np.array([[0.5, 0, 0], [0, 0.5, 0], [0, 0, 1]], dtype=np.float32)

y1 = warp(x, A, output_shape=(2*M, 2*N), order=0)
y2 = warp(x, A, output_shape=(2*M, 2*N), order=1)
y3 = warp(x, A, output_shape=(2*M, 2*N), order=2)

plt.figure(2)
plt.subplot(3,1,1)
plt.imshow(y1, clim=[0,255], cmap='gray')
plt.title('nearest')
plt.subplot(3,1,2)
plt.imshow(y2, clim=[0,255], cmap='gray')
plt.title('bilinear')
plt.subplot(3,1,3)
plt.imshow(y3, clim=[0,255], cmap='gray')
plt.title('bicubic')

# A è diversa da T. in particolare A = T[[1,0,2],:][:,[1,0,2]].T
```

```python
#traslazione
path_im = path+'lena.jpg'
x = np.float32(io.imread(path_im))

T = np.array([[1,0,0],[0,1,0],[50,100,1]])
A = T[[1,0,2],:][:,[1,0,2]].T

y = warp(x, A, order=1, cval=255) #cval modifica il livello di grigio esterno alla foto

plt.figure(1)
plt.subplot(1,2,1)
plt.imshow(x, clim=[0,255], cmap='gray')
plt.subplot(1,2,2)
plt.imshow(y, clim=[0,255], cmap='gray')
plt.show()
```

```python
#rotazione
path_im = path+'lena.jpg'
x = np.float32(io.imread(path_im))

theta = np.pi/6
T = np.array([[np.cos(theta),np.sin(theta),0],[-np.sin(theta),np.cos(theta),0],[0,0,1]])
A = T[[1,0,2],:][:,[1,0,2]].T

y = warp(x, A, order=1, cval=0) #cval modifica il livello di grigio esterno alla foto

plt.figure(1)
plt.subplot(1,2,1)
plt.imshow(x, clim=[0,255], cmap='gray')
plt.subplot(1,2,2)
plt.imshow(y, clim=[0,255], cmap='gray')
plt.show()
```
